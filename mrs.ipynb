```python
import pandas as pd

movies_path = "/content/drive/MyDrive/ML HUB/Movie Rec. Sys/tmdb_5000_movies.csv"
credits_path = "/content/drive/MyDrive/ML HUB/Movie Rec. Sys/tmdb_5000_credits.csv"

movies_df = pd.read_csv(movies_path)
credits_df = pd.read_csv(credits_path)

movies_df.head(), credits_df.head()
```

```python
# We'll merge the two DataFrames into one, called 'movies'.
movies = pd.merge(movies_df, credits_df, left_on='id', right_on='movie_id')

# Display the first few rows of the new, combined DataFrame.
movies.head()
```

Final list of features will be: genres, keywords, cast, crew, and overview

**Genres & Keywords Cleaning**

```python
import ast

# Example of how the genres column looks (we use iloc[0] to view the first row)
print(movies['genres'].iloc[0])
```

```python
def convert(obj):
    """
    Converts a JSON-formatted string of lists into a list of names.
    """
    # ast.literal_eval() converts the string into a Python list object
    try:
        L = [i['name'] for i in ast.literal_eval(obj)]
        return L
    except:
        # returns an empty list if there's an error (e.g., empty string)
        return []

# Apply this function to the 'genres' column
movies['genres'] = movies['genres'].apply(convert)

# Apply the same logic to the 'keywords' column since it has the same structure
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
print(movies['genres'].iloc[0])
print(movies['keywords'].iloc[0])
```

**Cleaning Cast**

```python
def get_top_3_cast(obj):
    """Extracts the names of the top 3 cast members."""
    # Convert string to list of dictionaries
    try:
        L = [i['name'] for i in ast.literal_eval(obj) if 'name' in i]
        return L[:3] # We only keep the first three names
    except:
        return []

movies['cast'] = movies['cast'].apply(get_top_3_cast)
```

**Cleaning Crew**

```python
def get_director(obj):
    """Extracts the name of the Director from the crew list."""
    # Convert string to list of dictionaries
    try:
        L = ast.literal_eval(obj)
        for i in L:
            if i['job'] == 'Director':
                return i['name']
        return None # Return None if the Director is not found
    except:
        return None

movies['director'] = movies['crew'].apply(get_director)
```

**Cleaning for Unique Tags**

```python
# Function to apply lowercase and remove spaces from elements in a list
def clean_data(x):
    if isinstance(x, list):
        # We process each item in the list (genres, keywords, cast)
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        # This handles single strings like 'director' and 'overview'
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

# Apply the cleaning function to our list-based columns
for col in ['genres', 'keywords', 'cast']:
    movies[col] = movies[col].apply(clean_data)

# Apply the cleaning to the director column
movies['director'] = movies['director'].apply(clean_data)
```

**Tag Soup**

```python
def create_soup(features):
    # Combine genres, keywords, cast (up to 3), and director into a single string.
    # Note: overview is already a string, but the others are lists, so we join them with spaces.

    # Overview needs a check for missing values (NaN), so we'll use a safer approach for the final soup.

    # We will join the lists with a space and then append the overview

    # The 'overview' column is the main plot summary (it needs its NaN values replaced)
    overview_text = features['overview'] if isinstance(features['overview'], str) else ""

    # Combine everything else
    return ' '.join(features['keywords']) + ' ' + \
           ' '.join(features['cast']) + ' ' + \
           ' '.join(features['genres']) + ' ' + \
           features['director'] + ' ' + \
           overview_text

# Fill any missing overview values with an empty string
movies['overview'] = movies['overview'].fillna('')

# Apply the function to create the 'soup' (our final combined tags)
movies['soup'] = movies.apply(create_soup, axis=1)

# Display the final soup for the first movie
print(movies['soup'].iloc[0])
```

**Vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer.
# We'll limit the features (words) to the top 5000 most frequent ones for efficiency.
cv = CountVectorizer(stop_words='english', max_features=5000)

# Fit the data (learn the vocabulary) and transform the data (create the matrix)
count_matrix = cv.fit_transform(movies['soup'])

# Display the shape of the matrix (rows = movies, columns = features/words)
print(f"Shape of the Count Matrix: {count_matrix.shape}")
```

**Calculating Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
# The result is a matrix where [i, j] is the similarity between movie i and movie j
cosine_sim = cosine_similarity(count_matrix)

# Display the shape of the similarity matrix
print(f"Shape of the Similarity Matrix: {cosine_sim.shape}")
```

**Building the Recommender Function**

```python
indices = pd.Series(movies.index, index=movies['title_x']).drop_duplicates()
```

**Creating the Recommender Function**

```python
# The final recommendation function
def get_recommendations(title):
    """
    Takes a movie title, finds its most similar movies, and returns the top 10.
    """
    # 1. Get the index of the movie that matches the title
    # We use the corrected indices Series here
    try:
        idx = indices[title]
    except KeyError:
        print(f"Error: Movie '{title}' not found in the dataset.")
        return

    # 2. Get the pairwise similarity scores of all movies with that movie
    # enumerate converts the scores array into a list of (index, score) tuples
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 3. Sort the movies based on the similarity scores (descending)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 4. Get the scores for the top 10 most similar movies
    # We slice [1:11] to skip the first element (which is the movie itself)
    sim_scores = sim_scores[1:11]

    # 5. Get the movie indices from the sorted list
    movie_indices = [i[0] for i in sim_scores]

    # 6. Return the top 10 movie titles
    # We use the corrected 'title_x' column
    return movies['title_x'].iloc[movie_indices]
```

**Testing**

```python
# Code for a true user prompt in a script/notebook
movie_choice = input("Enter a movie title to get recommendations: ")
recommendations = get_recommendations(movie_choice)
print(recommendations)
```
